```python
import re
import requests
import bs4

data = []
res = requests.get("https://www.toolify.ai/category", timeout=50)
res.raise_for_status()

soup = bs4.BeautifulSoup(res.text, 'html.parser')

table = soup.find_all('div', id=re.compile(r'^group-'))
for row in table:
    title = row.find('h3').get_text(strip=True)
    sub_categories = []

    for sub_row in row.find_all('a', class_='go-category-link'):
        # Extract the href attribute for the link
        href = sub_row.get('href', '')
        
        # Create full URL
        full_link = f"https://www.toolify.ai{href}" if href.startswith('/') else href
        
        name_span = sub_row.find('span', class_='text-base text-gray-1000 flex-1 w-0 truncate font-medium')
        if name_span:
            name = name_span.get_text(strip=True)
            sub_categories.append({
                    'name': name,
                    'link': full_link,
            })
    data.append(
        {
            'category': title,
            'sub_categories': sub_categories
        }
    )
print(data)
```

```python
import re
import requests
import bs4
sub_data = []
subcategory_url ='https://www.toolify.ai/category/ai-newsletter-generator'
try:
    print(f"Fetching subcategory page: {subcategory_url}")
    res = requests.get(subcategory_url, timeout=50)
    res.raise_for_status()
    print(f"Successfully fetched subcategory page, status code: {res.status_code}")
except requests.exceptions.RequestException as e:
    print(f"Failed to fetch subcategory page: {e}")


try:
    print("Starting HTML parsing for subcategory")
    soup = bs4.BeautifulSoup(res.text, 'html.parser')
    table = soup.find_all('div', class_='tool-item mb-10 border border-2 rounded-md overflow-hidden')
    print(f"Found {len(table)} tool items in subcategory")

    for i, row in enumerate(table):
        print(f"Processing tool item {i + 1}/{len(table)}")
        try:
            row_class = row.find('div', class_='tool-card bg-white p-6 flex gap-4 items-center')
            agent = row_class.find('div',
                                    class_='card-text-content flex flex-col justify-between flex-1 h-full w-0')

            name = agent.find('a').find('h2').get_text(strip=True)
            description = agent.find('p').get_text(strip=True)
            print(agent.find('a'))
            visit_btn_parent = row.find('div', class_='visit-btn')
            if visit_btn_parent:
                visit_link_tag = visit_btn_parent.find_parent('a')  # Get the <a> that wraps the button
                if visit_link_tag and visit_link_tag.has_attr('href'):
                    visit_url = visit_link_tag['href']
                    full_url = visit_url

            tool_data = {
                "name": name,
                "description": description,
                "url": full_url,
                "source": "https://www.toolify.ai",
                "category": "category_name"
            }

            sub_data.append(tool_data)
            print(f"Successfully extracted tool: {name}")

            print(f"Saved tool data for: {name}")
            print(sub_data)

        except AttributeError as e:
            print(f"Could not parse tool item {i + 1}: {e}")

except Exception as e:
    print(f"Unexpected parsing error in subcategory scraping: {e}")
```
